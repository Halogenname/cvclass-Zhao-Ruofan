```python
import pycolmap
from pathlib import Path
```

```python
image_dir = Path(r"D:\cvclass\images")
output_dir = Path(r"D:\cvclass")
```

```python
output_dir.mkdir(exist_ok=True)
sparse_dir = output_dir / "sparse"
sparse_dir.mkdir(exist_ok=True)
mvs_dir = output_dir / "mvs"
mvs_dir.mkdir(exist_ok=True)
database_path = output_dir / "database.db"
```

```python
print("1. 提取特征...")
pycolmap.extract_features(database_path, image_dir)
```

```python
print("2. 匹配特征...")
pycolmap.match_exhaustive(database_path)
```

```python
print("3. 稀疏重建...")
maps = pycolmap.incremental_mapping(database_path, image_dir, sparse_dir)
```

```python
model = maps[0]
model.write(sparse_dir)
print(f"稀疏重建完成: {len(maps[0].images)} 张图像, {len(maps[0].points3D)} 个3D点")
```

```python
sparse_ply = output_dir / "sparse_points.ply"
model.export_PLY(sparse_ply)
print("✓ 导出点云完成")
```
