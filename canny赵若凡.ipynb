```python
import cv2
```

```python
img = cv2.imread(r"C:\picture.png",0)
```

```python
img_blur = cv2.GaussianBlur(img,(5,5),1.4)
canny = cv2.Canny(img,100,200)
```

```python
cv2.imshow("原始图像",img)
cv2.imshow('canny算法生成结果',canny)
cv2.imwrite('canny.jpg',canny)
```
