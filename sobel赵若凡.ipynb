```python
import cv2
import numpy as np

```

```python
img = cv2.imread(r"C:\picture.png", 0)
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel = np.sqrt(sobel_x**2 + sobel_y**2)
sobel = np.uint8(sobel)
```

```python
cv2.imshow('原始图像', img)
cv2.imshow('Sobel的X计算', np.uint8(np.absolute(sobel_x)))
cv2.imshow('Sobel的Y计算', np.uint8(np.absolute(sobel_y)))
cv2.imshow('Sobel结果', sobel)
```
