```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r"C:\picture.png", 0)
img_blur = cv2.GaussianBlur(img, (3, 3), 0)
```

```python
laplacian = cv2.Laplacian(img_blur, cv2.CV_64F, ksize=3)
laplacian = np.uint8(np.absolute(laplacian))
```

```python
cv2.imshow('原始图像', img)
cv2.imshow('Laplacian结果', laplacian)
cv2.imwrite('laplacian_result.jpg', laplacian)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
